```python
import zipfile
import pandas as pd
import numpy as np

# Unzip the file
with zipfile.ZipFile("/content/archive.zip", "r") as z:
    z.extractall("/content/")

# Load data
df = pd.read_csv("/content/train.csv")
print("Shape:", df.shape)
print("\nMissing Values:\n", df.isnull().sum())
print("\nDuplicates:", df.duplicated().sum())

# Fix missing Postal Codes
df["Postal Code"] = df["Postal Code"].fillna(0).astype(int)

# Remove duplicates
df = df.drop_duplicates()

# Fix date formats
df["Order Date"] = pd.to_datetime(df["Order Date"], dayfirst=True)
df["Ship Date"] = pd.to_datetime(df["Ship Date"], dayfirst=True)

# Add new columns
df["Order Year"] = df["Order Date"].dt.year
df["Order Month"] = df["Order Date"].dt.month
df["Shipping Days"] = (df["Ship Date"] - df["Order Date"]).dt.days

# Standardize text
for col in ["Ship Mode","Segment","Country","City","State","Region","Category","Sub-Category"]:
    df[col] = df[col].str.strip().str.title()

# Save cleaned file
df.to_csv("/content/superstore_cleaned.csv", index=False)
print("\n✅ Cleaning Done!")
print(df.describe())
```
